# Problem 1

$$f(x'')=\frac{1}{2}(1-cos(x''))$$

Filter: $$x'' \in [x-\frac{\pi}{2}, x+\frac{\pi}{2}]$$


```python
import numpy as np

f = lambda x: 1 / 2 * (1 - np.cos(x))
```

---
#### i)
$$\overline{f}(x)=\frac{1}{2}-\frac{cos(x)}{\pi}$$

```python
f_bar = lambda x: 1 / 2 - np.cos(x) / np.pi
```

---
#### ii)

```python
import matplotlib.pyplot as plt

points = np.linspace(start=-5 * np.pi, stop=5 * np.pi, num=100)

plt.plot(points, f(points), label='f(x)')
plt.plot(points, f_bar(points), label='f_bar(x)')
plt.legend()
plt.show()
```

---
#### iii)

$$f'' \equiv f(x'') - \overline{f}(x'')$$

---
#### iv)

```python
f_prime_prime_bar = lambda x: -((np.pi - 2) / np.square(np.pi))*np.cos(x)

plt.plot(points, f(points), label='f(x)')
plt.plot(points, f_bar(points), label='f_bar(x)')
plt.plot(points, f_prime_prime_bar(points), label='f_prime_prime_bar(x)')
plt.legend()
plt.show()
```
